```python
from sklearn.datasets import fetch_openml
```

```python
data = fetch_openml('adult',as_frame=True).frame
data.head()
```

```python
data.occupation.value_counts()
```

```python
print(data.columns)
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

occupation_encoded = encoder.fit_transform(
    data[['occupation']]
)

encoded_df = pd.DataFrame(
    occupation_encoded,
    columns=encoder.get_feature_names_out(['occupation']),
    index=data.index
    )

data= data.drop(columns=['occupation'])
data=pd.concat([data,encoded_df],axis=1)

```

# The purpose of PCA is to reduce the number of features in your data while keeping the most important patterns,
#  making it easier to visualize and faster to process.

```python
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
x,y = fetch_openml('mnist_784',return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=10)
```

```python
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
pca = PCA(n_components=10)

x_train_reduced = pca.fit_transform(x_train_scaled)
x_test_reduced = pca.transform(x_test_scaled)
```

```python
x_train.shape
```

```python
x_train_reduced.shape # reduced from 784 to only 10
```

```python
clf = LogisticRegression(max_iter=100)
```

```python
clf.fit(x_train_scaled,y_train)

print('-------------------------------')
print(clf.score(x_test_scaled,y_test))
print('-------------------------------')
```

```python
clf.fit(x_train_reduced, y_train)

print('-------------------------------')
print(clf.score(x_test_reduced,y_test))
print('-------------------------------')
```

```python
clf.predict(x_test_scaled)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
x,y= load_breast_cancer(return_X_y=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=10)
```

```python
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
clf = KNeighborsClassifier()

clf.fit(x_train_scaled, y_train)
```

```python
clf.score(x_test_scaled,y_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, r2_score
```

```python
y_pred = clf.predict(x_test_scaled)
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
