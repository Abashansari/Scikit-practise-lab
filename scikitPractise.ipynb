```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder 
```

```python
df = fetch_openml('car',as_frame=True).frame
df
```

```python
# convert into int/numbers
columns_to_encode=["lug_boot","safety"]

encoder = OrdinalEncoder(
    categories=[
        ['small','med','big'],
        ['low','med','high']
    ]
)

df[columns_to_encode] = encoder.fit_transform(df[columns_to_encode])
```

```python
df.head()
```

```python
# to reverse it 
encoder.inverse_transform(df[columns_to_encode])
```

```python
from sklearn.datasets import fetch_openml
```

```python
data = fetch_openml('adult',as_frame=True).frame
data.head()
```

```python
data.occupation.value_counts()
```

```python
print(data.columns)
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

occupation_encoded = encoder.fit_transform(
    data[['occupation']]
)

encoded_df = pd.DataFrame(
    occupation_encoded,
    columns=encoder.get_feature_names_out(['occupation']),
    index=data.index
    )

data= data.drop(columns=['occupation'])
data=pd.concat([data,encoded_df],axis=1)

```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data=load_breast_cancer()
data
```

```python
x,y = load_breast_cancer(return_X_y=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier()

clf.fit(x_train_scaled, y_train)
```

```python
clf.score(x_test_scaled,y_test)
```

```python
single_instance = x_test_scaled[1]
single_instance
```

```python
clf.predict([single_instance])
```

```python
y_test[1]
```

```python
clf.predict_proba([single_instance])
```
